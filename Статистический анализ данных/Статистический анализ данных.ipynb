# Статистический анализ данных

Компания популярного сервиса аренды самокатов GoFast располагает данными о пользователях из нескольких городов и оказанных им услугах аренды самокатов. Чтобы совершать поездки по городу, клиенты сервиса GoFast пользуются мобильным приложением. Сервисом можно пользоваться без подписки с повышенными тарифами и с подпиской по льготным тариам.

## Описание данных.

Данные представлены в виде трёх таблиц:
1) Пользователи — users_go.csv -  содержит уникальный идентификатор пользователя, имя пользователя, возраст, город, тип подписки (free, ultra)

2) Поездки — rides_go.csv - содержит уникальный идентификатор пользователя, расстояние, которое пользователь проехал в текущей сессии (в метрах), продолжительность сессии (в минутах), дата совершения поездки

3) Подписки — subscriptions_go.csv - тип подписки, стоимость одной минуты поездки по данной подписке, стоимость начала поездки, стоимость ежемесячного платежа


## Цели и задачи
Целью проекта ставится анализ имеющихся данных для роста бизнеса: менеджеры сервиса хотят увеличить количество пользователей с платной подпиской с помощью неоторых акций по привлечению клиентов.

Необходимо проверить гипотезы:
- тратят ли пользователи с платной подпиской больше времени на поездки? Если да, то пользователи с платной подпиской могут быть «выгоднее» для компании;
- можно ли сказать, что среднее расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров (оптимальное расстояние с точки зрения износа самоката);
- будет ли помесячная выручка от пользователей с платной подпиской по месяцам выше, чем выручка от пользователей без подписки;
- уточнить метод проверки гипотезы о том, снизится ли количество обращений в техподдержку после обновления серверов, с которыми взаимодействует мобильное приложение.

Дополнительные задачи:
- отделу маркетинга GoFast поставили задачу: провести акцию с раздачей промокодов на один бесплатный месяц подписки, в рамках которой как минимум 100 существующих клиентов должны продлить эту подписку. По завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соответствующий платёж. Из результатов проводимой ранее подобной акции известно, что после бесплатного пробного периода подписку продлевают 10% пользователей. Необходимо определить, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно 5 %;

- отдел маркетинга рассылает клиентам push-уведомления в мобильном приложении, которые клиенты могут открыть или не открывать. Известно, что уведомления открывают около 40 % получивших клиентов. Отдел планирует разослать 1 млн уведомлений. Необходимо представить примерный график распределения и оцените вероятность того, что уведомление откроют не более 399,5 тыс. пользователей.

```python
#Сгруппируем исходную таблицу пользователей по типу подписки и подсчитаем их количество:
users_subscription_type = users.groupby('subscription_type').agg({'user_id':'count'}).reset_index()
print('Количество пользователей с платной подпиской (ultra) и бесплатной подпиской (free):')
display(users_subscription_type)
```

## Проверка гипотез

Продакт-менеджеры сервиса хотят увеличить количество пользователей с подпиской. Для этого будут проводиться различные акции, но сначала нужно уточнить результативность нескольких гипотез.

### Важно понять, тратят ли пользователи с подпиской больше времени на поездки? 

Если да, то пользователи с подпиской могут быть «выгоднее» для компании. 

Сформулируем гипотезы.

**Н0**:
Средняя продолжительность поездок с платной подпиской и с бесплатной равны.

**Н1**:
Средняя продолжительность поездок с платной подпиской больше средней продолжительности поездок с бесплатной подпиской.

<div class="alert alert-block alert-success">
<b>✅Комментарий ревьюера:</b> Здорово, что формулируешь гипотезы перед проверкой!</div>

```python
print('Средняя продолжительность поездок с платной подпиской равно', np.mean(data_ultra['duration']))
print('Средняя продолжительность поездок с бесплатной подпиской равно', np.mean(data_free['duration']))

# Так мы имеем две независимые выборки, проверку следует сделать методом st.ttest_ind:
results = st.ttest_ind(data_ultra['duration'],
                       data_free['duration'],
                       equal_var=True,
                       alternative='greater')
# выведем значение верояности ошибки:
print('Значение p-value', results.pvalue)
# выведем результат проверки гипотезы:
alpha = 0.05
if results.pvalue < alpha:
    print('Нулевую гипотезу следует отвергнуть.')
else:
    print('Не следует отвергать нулевую гипотезу.')
```

**Вывод**

Нулевая гипотезу о равенстве длительности поездок при разном типе подписки следует отвергнуть. Анализ даёт основания утверждать, что средняя продолжительность поездки с подпиской больше продолжительности поездок без подписки при установленном уровне значимости в 5 %. Это подтверждается ранее полученными числовыми характеристиками этих групп данных о том, что средние значения отличаются примерно на минуту.

### Соответствует ли среднее расстояние одной поездки оптимальному с точки зрения износа самоката величиной в 3130м. 

Можно ли сказать, что среднее расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров?

Сформулируем гипотезы.

**Н0**:
Среднее расстояние поездок с платной подпиской равно оптимальному расстоянию в 3130 метров.

**Н1**:
Среднее расстояние поездок с платной подпиской больше оптимального расстояния в 3130 метров.

```python
# Выведем среднюю дистанцию поездок с платной подпиской:
print('Среднее расстояние поездок с платной подпиской равно', np.mean(data_ultra['distance']))
# Так мы имеем одну выборку, проверку следует сделать методом st.ttest_1samp
# Зададим внешние параметры для теста:
distance = 3130
alpha = 0.05

results = st.ttest_1samp(data_ultra['distance'], distance, alternative='greater')
# выведем значение верояности ошибки:
print('Значение p-value', results.pvalue)

# выведем результат проверки гипотезы:
if results.pvalue < alpha:
    print('Нулевую гипотезу следует отвергнуть.')
else:
    print('Не следует отвергать нулевую гипотезу.')
```

**Вывод**

Согласно результатам проверки гипотезы при установленном уровне значимости в 5%, среднее значение дальности поездки пользователей с подпиской не превышает оптимальное с точки зрения износа самоката расстояние в 3130 м. Нулевую гипотезу отвергнуть не удалось, а расчитанное среднее арифметическое в 3115,45м подтверждает результат.

### Будет ли помесячная выручка от пользователей с подпиской по месяцам выше, чем выручка от пользователей без подписки?

Сформулируем гипотезы.

**Н0**:
Помесячная средняя выручка от пользователей с платной подпиской равна помесячной средней выручке от пользователей с бесплатной подпиской.

**Н1**:
Помесячная средняя выручка от пользователей с платной подпиской больше помесячной средней выручки от пользователей с бесплатной подпиской.

```python
print('Средняя выручка от пользователей с платной подпиской равна', 
      np.mean(revenue_users_by_months.loc[revenue_users_by_months['subscription_fee']>0, 'revenue'])
     )
print('Средняя выручка от пользователей с бесплатной подпиской', 
      np.mean(revenue_users_by_months.loc[revenue_users_by_months['subscription_fee']==0, 'revenue'])
     )

# Так мы имеем две независимые выборки, проверку следует сделать методом st.ttest_ind:
results = st.ttest_ind(revenue_users_by_months.loc[revenue_users_by_months['subscription_fee']>0, 'revenue'],
                       revenue_users_by_months.loc[revenue_users_by_months['subscription_fee']==0, 'revenue'],
                       equal_var=True,
                       alternative='greater')
# выведем значение верояности ошибки:
print('Значение p-value', results.pvalue)
# выведем результат проверки гипотезы:
alpha = 0.05
if results.pvalue < alpha:
    print('Нулевую гипотезу следует отвергнуть.')
else:
    print('Не следует отвергать нулевую гипотезу.')
```

**Вывод**

Согласно проведённой проверки, следует ожидать, что средняя выручка от пользователей с платной подпиской больше помесячной средней выручки от пользователей с бесплатной подпиской. Вычисленные средние значения для обеих групп пользователей соответствуют этому утверждению.

### Снизится ли количество обращений в техподдержку после обновления серверов, с которыми взаимодействует мобильное приложение? Подобрать метод проверки гипотез.

Сформулируем гипотезы.

**Н0**:
Количество обращений в техподдержку до и после обновления серверов не изменилось.

**Н1**:
Количество обращений в техподдержку после обновления серверов уменьшилось.

Сформулированная ситуация предполагает, что выборки зависимы. Это значит, что рассматривается одна и та же генеральная совокупность (данные об обращениях в техподдержку) и некоторая переменная (количество обращений) измерена два раза для одних и тех же объектов наблюдения — до и после какого-то изменения (обновление серверов, с которыми взаимодействует мобильное приложение). Поэтому выборки обязательно должны быть одного размера и согласованы друг с другом.

### Расчет минимального количества промокодов в акции на один бесплатный месяц подписки.
**Задача**: 
при проведении акции с раздачей промокодов на один бесплатный месяц подписки, как минимум 100 существующих клиентов должны продлить эту подписку. По завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соответствующий платёж. Из результатов проводимой ранее подобной акции известно, что после бесплатного пробного периода подписку продлевают 10% пользователей. Необходимо определить, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно 5 %.


```python
p = 0.1 # вероятность продления подписки из предыдущих акций
n = 1000 # предположим рассылку 1000 промокодов
x = 100 # количество желаемых клиентов

# на основании биномиального распределения вычислим количество промокодов с желаемой вероятностью выполнения плана
while binom.cdf(x, n, p)>0.05:
    n += 1

# математическое ожидание нормального распределения зададим равным биномиальному:
mu = n*p
# стандартное отклонение нормального распределения зададим равным биномиальному:
sigma = (n * p * (1 - p))**0.5
# зададим интервал нормального распределения с максимально возможным попаданием данных и шагом в единицу:
arange = np.arange((mu - 4*sigma),(mu + 4*sigma),1)
# построим график нормального распределения:
plt.plot(arange, st.norm.pdf(arange, mu, sigma))
plt.axvline(x=x, color='green') # отметим желаемое количество клиентов
plt.title(f'График распределения вероятного количества клиентов \n при рассылке {n} промокодов')
plt.show()
print('При раздаче промокодов в количестве', n, '- вероятность не выполнить план равна 5%')
```

**Вывод**

Для уверенного выполнения плана по привлечению клиентов с помощью промокода на бесплатный пробный месяц подписки ultra необходимо отправить не менее 1172 промокодов. С вероятностью продления подписки в 10% (согласно ранее проводимым подобным акциям), как минимум 100 существующих клиентов должны продлить эту подписку. Вероятность невыполнения плана привлечения в размере 100 клиентов составит 5%.

### Оценить вероятность открытия push-уведомления и визуализировать распределение.
**Задача**: отдел маркетинга рассылает клиентам push-уведомления в мобильном приложении, которые клиенты могут открыть или не открыть. Известно, что уведомления открывают около 40 % получивших клиентов. Отдел планирует разослать 1 млн уведомлений. Необходимо представить примерный график распределения и оценить вероятность того, что уведомление откроют не более 399,5 тыс. пользователей.

```python
# параметры биномиального распределения:
n = 1000000 # количество push-уведомления
p = 0.4 # вероятность открытия push-уведомления
x = 399500 # необходимое количество открытий push-уведомления

# математическое ожидание нормального распределения зададим равным биномиальному:
mu = n*p
# стандартное отклонение нормального распределения зададим равным биномиальному:
sigma = (n * p * (1 - p))**0.5

# задаём нормальное распределение с указанными параметрами:
distr = st.norm(mu, sigma) 

# считаем вероятность получить значение, равное количеству открытий х или меньше:
result = distr.cdf(x)

# зададим интервал нормального распределения с максимально возможным количеством данных и шагом в единицу:
arange = np.arange((mu - 4*sigma),(mu + 4*sigma),1)
# построим график распределения:
plt.plot(arange, st.norm.pdf(arange, mu, sigma))
plt.xticks(rotation=70)
plt.axvline(x=x, color='green') # отметим желаемое значение открытий
plt.title(f'График распределения вероятности открытия \n push-уведомлений')
plt.show()

print(f'Вероятность открытия push-уведомления: {(result*100).round(2)}%')
```

**Вывод**

При рассылке 1 миллиона push-уведомлений менее 399,5 тысяч из них будет открыто с общей вероятностью 15,37%.

- клиенты моложе 20 и старше 30 лет чаще отдают предпочтение платной подписке, в то время как клиенты от 20 до 30 лет больше склоняются к бесплатному пользованию приложением;

- длительность поездок распределена нормально, медианное значение 18 мин почти полностью совпадает со средним 18,3 мин. Наблюдающийся отдельный пик околонулевых значений характерен для наличия в выборке выбрасов, отвечающих данным промо-акций;

- в распределении дальности поездок наблюдается два основных пика: малый на расстоянии от 500 до 1000 м и большой в основной массе поездок на расстоянии от 2900м до 3500м. Основная часть распределения (при расстоянии более 1500 м) выглядит нормально, но малый пик отбрасывать не стоит. Медианное значение расстояния в 3133.61 м довольно близко к среднему в 3070.66 м. Наличие двух пиков на гистограмме распределения может говорить о смешении двух распределений. То есть возможно вмешательство в статистику общей выборки небольшого количества данных с проводимыми в 2021 году акциями по привлечению клиентов, связанными с пробными периодами или льготными дистанциями;

- распределение скорости скорее правостороеннее, что подтверждается чем, что среднее значение 197,38 м/мин больше медианы 166,18 м/мин. Кроме того, расчёт по данным длительности и дальности поездки фиксирует максимальную скорость 7211.01 м/мин. Это связано с наличием выбросов минимального времени при проведении промо-акций.

Изучение данных пользователей и поездок с разными видами подписки показало, что этот параметр имеет значение и сказывается не только на характеристиках пользователей, но и на параметрах поездок:

- медианные значения дистанции поездок больше почти на 34м при платной подписке;

- разница между медианным и средним значением в случае платной подписки почти в два раза меньше разницы этих величин при бесплатной подписке, это подтвержается ощутимой разницей стандартного отклонения и говорит о меньшем разбросе данных платной подписки;

- наличие платной подписки увеличивает минимальные и средние значения времени поезки примено на 1 минуту (с 18 мин для бесплатной подписки до 19 мин для платной),

- при платной подписке зафиксированы длительность поездки становится больше, особенно в максимальных значениях (35 мин при бесплатной подписке против 41 мин при платной).

Сформулированная ситуация о проведении обновлений серверов предполагает, что выборки зависимы. Это значит, что рассматривается одна и та же генеральная совокупность (данные об обращениях в техподдержку) и некоторая переменная (количество обращений) измерена два раза для одних и тех же объектов наблюдения — до и после какого-то изменения (обновление серверов, с которыми взаимодействует мобильное приложение). Поэтому выборки обязательно должны быть одного размера и согласованы друг с другом.
